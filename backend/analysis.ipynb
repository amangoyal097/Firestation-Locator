```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
with open("data.json","r") as f:
    data = json.load(f)
```

```python
plt.xlabel("Point number")
plt.text(0,4.7,"Minimum Time %0.2f minutes" % (np.min(data["optimal_times"])))
plt.text(0,4.5,"Maximum Time %0.2f minutes" % (np.max(data["optimal_times"])))
plt.text(0,4.3,"Average Time %0.2f minutes" % (np.average(data["optimal_times"])))
plt.ylabel("Time (in minutes)")
plt.title("Time taken by the nearest fire station")
plt.xticks([i + 1 for i in range(len(data["optimal_times"]))])
plt.plot([i + 1 for i in range(len(data["optimal_times"]))],data["optimal_times"],marker='o')
```

```python
plt.xlabel("Point number")
plt.ylabel("Time (in minutes")
hour_arr = [i + 1 for i in range(24)]
point_arr = [i + 1 for i in range(len(data["optimal_times"]))]
actual_average_times = []
actual_min_times = []
actual_max_times = []
for arr in data["actual_times"]:
    actual_average_times.append(np.average(arr))
    actual_min_times.append(np.min(arr))
    actual_max_times.append(np.max(arr))
plt.plot(point_arr, actual_max_times,label="Maximum Times",marker='o')
plt.plot(point_arr, actual_average_times,label="Average Times",marker='o')
plt.plot(point_arr,actual_min_times,label="Minium Times",marker='o')
plt.xticks(point_arr)
plt.legend()
plt.title("Actual Times")
```

```python
max_diff = []
for i in range((len(point_arr))):
    maxi = 0
    arr = data["actual_times"][i]
    for val in arr:
        maxi = max(maxi, val - np.min(arr))
    max_diff.append(maxi)
plt.plot(point_arr,max_diff,marker='o')
plt.xlabel("Point number")
plt.ylabel("Time (in minutes")
plt.xticks(point_arr)
plt.title("Maximum Difference between actual time and optimal time")
```
